```python

import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,1,4,7,7,6.5,6.5,4,1.5,1.5,1])
y=np.array([2,10,6,10,2,2,8.5,5.2,8.5,2,2])
plt.plot(x,y,ls='-',color='m')
plt.show()

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.plot(x,y,ls='-',color='k')
plt.show()

x=np.array([1,3,5,6])
y=np.array([2,4,6,8])
plt.plot(x,y,ls='-',color='k')
plt.show()
```
